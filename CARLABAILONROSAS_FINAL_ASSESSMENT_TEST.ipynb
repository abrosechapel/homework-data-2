```python
# 1. load  sales_reps_data.csv file to a dataframe
import pandas as pd

data=pd.read_csv('sales_reps_data.csv')
# 2. Prepare a new subset from the original data, aggregating sales records by sales representative and summing up
# their total amount of sales (check sales_resp_data_aggregated.csv file to see what needs to be achieved)

"""
HINT: How to rename a column? 

 - sometimes when you group by column A and sum up colum B values, you lose the name of column B
 - below is the example of how to name the column B after group by
 - it is the equivalent of SQL: SELECT columnA, sum(columnB) as 'Total'

Example:
  new_df = <your groupby and sum code goes here>.reset_index(name ='total_amount')

"""

new_sales=data.groupby('sales_rep')['total'].sum().reset_index(name ='total_amount')

new_sales.to_csv('salesbyrep.csv')

new_sales
```

```python
# 3. Using matplotlib create a Pie plot to show proportion of sales contribution to the total sales by each sales rep. 

import matplotlib.pyplot as plt


rep = new_sales.sales_rep.tolist()
total =new_sales.total_amount.tolist()

cols = ['c','g','y','b','m','brown','orange','skyblue','gold' ]


fig = plt.figure(figsize =(15, 8))
plt.title('Pie plot sales by sales rep', fontsize=20)

plt.pie(total, labels = rep,colors= cols, autopct='%1.2f%%')
 
plt.show()

```

The big contribution in sales has been done by the sales representative Kivell, followed by Parent and Jardine.

### Q1: What is Machine Leanring?
Machine learning is a field of computer science that teach a computer how to classify, predict and determine. The machine learning consist in create models to teach to the computer and the next process the program do it by itself. The algorithms  are trained with past information to prove their efficence with new data. An example can be predict if a movie are going to be sucessful or not, profitable or not, or the rate that is goign to have. There are four types: supervised learning
Unsupervised learning, semi-supervised learning and reinforcement learning.

### Q2: What are the key principles of 'Supervised' Machine learning?

In supervised machine learning there is a data that has been worked before and accoeding this results we want to the computer predicts the result. Like in the movie prediction example, the data of the movies has the result predicted then we use this to create the algorithm to predict according of this data.


### Q3: With regards to Machine Learning Model Fitting, what do we mean my Underfitting and Overfitting?

**Underfitting**, is an error that occurs when a statistical model or machine learning algorithm fails to capture the underlying trend in the data when the model or algorithm does not fit the data well enough. Here the model is very simplistic, insufficient to capture the complexities in the data. This occurs for the low variance but high bias. Also because our data is not enough or we didn't understand it well.

**Overfitting**, is a modeling error that occurs when a model is excessively complex, becuase it has many parameters relative to the number of observations or data points. The model is trained with so much data, that it learns from noise and inaccurate data inputs. The model is going to have poor predictive performance.

### Q4: Explain how do we ususally split data in preparation for the Machine Learning process?
The data is split in data test and training. Usually with 30 and 70 percent respectively(the percent may vary). The data training is the data that is going to work with machine learning to learn how to predict and the test is the data to use to prove the model.
