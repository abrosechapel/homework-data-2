### 1. GRAPH 1: â€œCreate a graph to find relationship between the sepal length and width."

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris_data = pd.read_csv("iris.csv")

fig = iris_data[iris_data.species=='setosa'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='orange', label='Setosa')
iris_data[iris_data.species=='versicolor'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='blue', label='versicolor',ax=fig)
iris_data[iris_data.species=='virginica'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='green', label='virginica', ax=fig)

fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("Sepal Length VS Width")
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```

<div style="text-align: justify">
In a global view of the graph the sepal length and sepal width don't have any relation because all the data points has a very high spread, that means that the data points are over all the graph. The setosa (the yellow points) seems different and far from the other two species and is concentrated in a upper left or top corner zone of the graph.If we review by iris species, there is a relation between sepal legnth and witdth of the Setosas,all the data points are concentrated in a section and if we review carefully it looks like a linear positive correlation, in comparision with the others the width values are high. The other two types of iris are entangle because they have similar values and if we look them separately it is clear that there is any relationship between the sepal width and sepal legnth of virginica, also there isn't a relation between sepal lenght and sepal width of versicolors.

### GRAPH 2: " Create a joinplot to describe individual distributions on the same plot between Sepal length and Sepal width. "


```python
#hexagonal bin plot

fig = sns.jointplot(x='sepal_length', y='sepal_width', kind='hex', color='blue', data=iris_data)
plt.show()
```

<p style='text-align: justify;'>


The joinplot function generates the hexagonal bin plot, this show three graphs. The graphs from the side are the distribution of two variables and the graphic from the center is an scatter plot with hexagonal bins from the two variables.
  
</p>
<p style='text-align: justify;'>

The histogram on the top shows the distribution of the variable at the x-axis or independent variable (X) and the histogram to the right shows the distribution of the variable at the y-axis or dependent variable (Y). The x-axis histogram (in the top) is bimodal, asymetric and it is a little right skewed. The y-axis histogram (in the right) is concentrated in the center and it is a asymetric.
</p>
<p style='text-align: justify;'>
The center plot displays a bivariate graph which shows how the dependent variable (Y) varies with the independent variable (X).  The bivariate analysis explores the relationship between two variables and also describes the strength of their relationship by  grouping the values by a range of values. This is known as binning. The entire plot is divided into a grid of bins, we can use grids of different shapes in this graph the bin is an hexagon. Every bin in the plot represents an interval and it contains all the data points within the defined x and y value ranges. The data points are counted according the bin that they belongs then the hexagons are colored based on a color gradient. Darker colors indicate that the data points are concentrated in that region the color gradient decrease when the data points decrease until it become white this indicates that there is no data points. So the plot show that there isn't any relation between the two variables and also show darker bins in the x-axis values to 5.5 from 6.0 and y values to 2.5 from 3.0, that is the zone where the values of versicolor and virginica are entwined. Also there are other darker heaxagon beetween the x value five and y value 3.5 that is the zone of virginica.
